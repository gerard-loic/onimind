| v8 | CNN | Conv(128) + 10 blocs residuels (Conv+Conv+BN+ReLU) | Conv(16)+Dropout+Dense(128)| Conv(64)+BN+Droupout+Conv(52) | Issu de l'architecture de v6, mais avec 10 couches de blocs résiduels (plus profond) | Abandonné. Trop lourd. |
| v9 | Dense | Dense(512)+BN+ReLU+Dropout+Dense(512)+BN+ReLU+Dropout+Dense(256)+BN+ReLU+Dropout | Dense(128)+ReLU+Dropout+Dense(64)+ReLU | Dense(512)+ReLU+Dropout | Architecture identique à v7, sur réseau dense, mais avec une couche de plus dans la tête de politique | Base du modèle Shigemori | 
| v10 | Dense | Dense(128)+ReLU+Dense(128)+ReLU+Dense(64)+ReLU+Skip  |Dense(64)+ReLU+Dense(32)+ReLU | Dense(64)+ReLU | Architecture sur réseau dense, sans LayerNormalization, sans Dropout, allégée avec skip connection | Base du modèle Sukoshi|
