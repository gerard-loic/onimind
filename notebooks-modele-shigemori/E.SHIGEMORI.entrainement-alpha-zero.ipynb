```python
import sys
sys.path.insert(0, '../onitama')

from dl_players_v9 import DensePlayer_v9
from alphazero import AlphaZeroTrainer
from players import LookAheadHeuristicPlayer

# ── Instanciation des joueurs ──────────────────────────────────────────────
p1 = DensePlayer_v9()
p2 = DensePlayer_v9()

p1.load_weights('../saved-models/Shigemori_rl_iter20.weights.h5')

# Adversaire de référence pour l'évaluation
eval_player = LookAheadHeuristicPlayer(heuristic_function="heuristic_defensive", max_depth=2)

trainer = AlphaZeroTrainer(
    player_p1=p1,
    player_p2=p2,
    n_games=50,
    n_simulations=400,
    temperature_moves=5,
    n_epochs=30,
    minibatch_size=256,
    replay_buffer_size=10000,
    learning_rate=3e-5,
    checkpoint_path='../saved-models/Shigemori_alphazero2',
    eval_player=eval_player,
    eval_games=200,   # 50 parties en P1 + 50 en P2
    eval_every=10,
    value_coef=2
)

history = trainer.train(n_iterations=100)
```
