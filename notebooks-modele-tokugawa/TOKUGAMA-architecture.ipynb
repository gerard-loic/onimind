# Architecture

La différence principale avec la v6 saute aux yeux : la tête de valeur est plus directe. En v5, on part du tronc directement avec un Conv2D(16, 1×1) sans passer par le value_adapter Conv2D(64, 3×3) qui existe dans la v6.

Concrètement, la v6 a introduit cet adaptateur pour deux raisons : donner à la tête de valeur ses propres features spatiales (extraites avec un noyau 3×3 au lieu de 1×1), et rendre le chargement de poids anciens plus propre (les couches de l'adaptateur sont absentes des anciens fichiers .h5, donc ignorées lors du load_weights avec skip_layers). En v5 ce mécanisme n'existe pas, d'où la méthode load_weights beaucoup plus simple qui se contente d'un appel direct.
