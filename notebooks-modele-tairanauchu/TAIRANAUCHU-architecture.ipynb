# Architecture

La v7 est une rupture architecturale nette par rapport aux v4/v5/v6 : zéro convolution, zéro bloc résiduel. Tout repose sur des couches denses.
Le changement fondamental commence dès l'entrée : le plateau 5×5×10 est immédiatement aplati en un vecteur (250,). On perd délibérément toute notion de spatialité — les relations entre cases voisines ne sont plus captées par des filtres glissants, mais devront être apprises implicitement par les couches denses. C'est exactement ce qui en fait un modèle de comparaison intéressant : est-ce que l'inductive bias spatial des CNN apporte vraiment quelque chose sur un plateau 5×5, ou un MLP suffisamment large s'en sort aussi bien ?

Le tronc 512 → 256 → 128 avec BN + Dropout(0.1) entre chaque couche est l'équivalent fonctionnel des blocs résiduels, en moins profond. Les têtes sont aussi plus directes : plus de Conv2D(52) + reshape pour la politique, juste un Dense(1300) ; et la tête de valeur se réduit à un seul Dense(64).
