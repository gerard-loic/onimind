```python
import sys
sys.path.append('../onitama/')

from dl_minimax import LookAheadDlPlayer
from players import LookAheadHeuristicPlayer
from dl_players_v4 import CNNPlayer_v4
from trainer import RegularDataTrainer
from game import GameSession
```

```python
PRETRAINED_WEIGHTS = '../saved-models/PPO-Model1.weights.h5'
OUTPUT_DATA = '../data/'
```

```python
cnn = CNNPlayer_v4()
cnn.load_weights(filepath=PRETRAINED_WEIGHTS)

teacher = LookAheadDlPlayer(max_depth=2, dl_player=cnn, n_best_moves=5)

p2 = LookAheadHeuristicPlayer(heuristic_function="heuristic_mobility", max_depth=3)
p3 = LookAheadHeuristicPlayer(heuristic_function="heuristic_aggressive", max_depth=2)

trainer = RegularDataTrainer(
    p1=p2, p2=p3,        # le teacher joue contre lui-même
    p1_record=True, p2_record=True,
    save_only_wins=True,           # on veut TOUS les coups, pas que les victoires
    x_file_destination=OUTPUT_DATA+'LookAheadHeuristic2Agressive-vs-LookAheadHeuristic3Mobility-states.pkl',
    y_file_destination=OUTPUT_DATA+'LookAheadHeuristic2Agressive-vs-LookAheadHeuristic3Mobility-actions.pkl',
    v_file_destination=OUTPUT_DATA+'LookAheadHeuristic2Agressive-vs-LookAheadHeuristic3Mobility-values.pkl',
)
session = GameSession(player_one=p2, player_two=p3, number_of_games=5000, trainer=trainer)
session.start()
```
