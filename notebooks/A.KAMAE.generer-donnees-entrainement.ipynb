```python
import sys
sys.path.append('../onitama/')

from dl_minimax import LookAheadDlPlayer
from players import LookAheadHeuristicPlayer
from dl_players_v4 import CNNPlayer_v4
from trainer import RegularDataTrainer
from game import GameSession
```

```python
OUTPUT_DATA = '../data/v6/'
pa_heuristic_function = 'heuristic_regular'
pb_heuristic_function = 'heuristic_positional'
```

```python


pa = LookAheadHeuristicPlayer(heuristic_function=pa_heuristic_function, max_depth=3)
pb = LookAheadHeuristicPlayer(heuristic_function=pb_heuristic_function, max_depth=2)

x_file = f"{OUTPUT_DATA}LookAheadHeuristic3{pa_heuristic_function}-vs-LookAheadHeuristic2{pb_heuristic_function}.states.pkl"
y_file = f"{OUTPUT_DATA}LookAheadHeuristic3{pa_heuristic_function}-vs-LookAheadHeuristic2{pb_heuristic_function}.labels.pkl"
v_file = f"{OUTPUT_DATA}LookAheadHeuristic3{pa_heuristic_function}-vs-LookAheadHeuristic2{pb_heuristic_function}.values.pkl"
m_file = f"{OUTPUT_DATA}LookAheadHeuristic3{pa_heuristic_function}-vs-LookAheadHeuristic2{pb_heuristic_function}.mask.pkl"

trainer = RegularDataTrainer(
    p1=pa, p2=pb,     
    p1_record=True, p2_record=True,
    save_only_wins=True,
    x_file_destination=x_file,
    y_file_destination=y_file,
    v_file_destination=v_file,
    m_file_destination=m_file
)
session = GameSession(player_one=pa, player_two=pb, number_of_games=5000, trainer=trainer)
session.start()
```
