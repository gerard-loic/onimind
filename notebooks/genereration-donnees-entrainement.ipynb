# Génération de données d'entraînement
Le but de ce notebook est de générer des données d'entraînement en faisant jouer les joueurs de référence les uns contre les autre. Les données d'entraînement servent à l'entraînement supervisé des modèles.

```python
#Chargement des librairies
import sys
sys.path.append('../onitama/')

from dl_minimax import LookAheadDlPlayer
from players import LookAheadHeuristicPlayer
from dl_players_v4 import CNNPlayer_v4
from trainer import RegularDataTrainer
from game import GameSession
```

```python
#Configuration
OUTPUT_DATA = '../data/new/'     #Dossier de sortie
number_of_games = 5000
variants = [
    {
        'p1' : LookAheadHeuristicPlayer(heuristic_function='heuristic_aggressive', max_depth=3),
        'p2' : LookAheadHeuristicPlayer(heuristic_function='heuristic_defensive', max_depth=2),
        'name' : 'LookAheadHeuristic3heuristic_aggressive-vs-LookAheadHeuristic2heuristic_defensive'
    },
    {
        'p1' : LookAheadHeuristicPlayer(heuristic_function='heuristic_aggressive', max_depth=3),
        'p2' : LookAheadHeuristicPlayer(heuristic_function='heuristic_regular', max_depth=2),
        'name' : 'LookAheadHeuristic3heuristic_aggressive-vs-LookAheadHeuristic2heuristic_regular'
    },
    {
        'p1' : LookAheadHeuristicPlayer(heuristic_function='heuristic_defensive', max_depth=3),
        'p2' : LookAheadHeuristicPlayer(heuristic_function='heuristic_aggressive', max_depth=2),
        'name' : 'LookAheadHeuristic3heuristic_defensive-vs-LookAheadHeuristic2heuristic_aggressive'
    },
    {
        'p1' : LookAheadHeuristicPlayer(heuristic_function='heuristic_defensive', max_depth=3),
        'p2' : LookAheadHeuristicPlayer(heuristic_function='heuristic_mobility', max_depth=2),
        'name' : 'LookAheadHeuristic3heuristic_defensive-vs-LookAheadHeuristic2heuristic_mobility'
    },
    {
        'p1' : LookAheadHeuristicPlayer(heuristic_function='heuristic_defensive', max_depth=3),
        'p2' : LookAheadHeuristicPlayer(heuristic_function='heuristic_positional', max_depth=2),
        'name' : 'LookAheadHeuristic3heuristic_defensive-vs-LookAheadHeuristic2heuristic_positional'
    },
    {
        'p1' : LookAheadHeuristicPlayer(heuristic_function='heuristic_defensive', max_depth=3),
        'p2' : LookAheadHeuristicPlayer(heuristic_function='heuristic_regular', max_depth=2),
        'name' : 'LookAheadHeuristic3heuristic_defensive-vs-LookAheadHeuristic2heuristic_regular'
    }
]
```

```python
#Génération des données

for variant in variants:
    p1 = variant["p1"]
    p2 = variant["p2"]

    x_file = f"{OUTPUT_DATA}{variant['name']}.states.pkl"
    y_file = f"{OUTPUT_DATA}{variant['name']}.labels.pkl"
    v_file = f"{OUTPUT_DATA}{variant['name']}.values.pkl"
    m_file = f"{OUTPUT_DATA}{variant['name']}.mask.pkl"

    trainer = RegularDataTrainer(
        p1=p1, p2=p2,     
        p1_record=True, p2_record=True,
        save_only_wins=True,
        x_file_destination=x_file,
        y_file_destination=y_file,
        v_file_destination=v_file,
        m_file_destination=m_file
    )
    session = GameSession(player_one=p1, player_two=p2, number_of_games=number_of_games, trainer=trainer)
    session.start()
```
