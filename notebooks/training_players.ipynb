```python
training_plan = [
        (
            LookAheadHeuristicPlayer(max_depth=3, heuristic_function="heuristic_aggressive"),
            LookAheadHeuristicPlayer(max_depth=3, heuristic_function="heuristic_defensive"),
            "agressive3-vs-defensive3test"
        )   
    ]

i = 1
for p1, p2, filename in training_plan:
    print(f"Training session {i}")
    trainer = RegularDataTrainer(
        p1=p1,
        p2=p2,
        p1_record=True,
        p2_record=True,
        save_only_wins=True,
        x_file_destination=f"{data_dir}/{filename}-states.pkl",
        y_file_destination=f"{data_dir}/{filename}-actions.pkl",
        override=True
    )
    gameSession = GameSession(player_one=p1, player_two=p2, number_of_games=10, trainer=trainer)
    gameSession.start()
    print(gameSession.getStats())
```
