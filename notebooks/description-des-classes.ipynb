# Description des classes

Le présent notebook via à donner une nomenclature des classes du module onitama, classes permettant : 
- De gérer l'environnement
- De gérér les joueurs de référence
- Ge gérer des parties et des sessions de partie
- De définir les joueurs neuronaux
- De gérer l'entraînement PPO
- de créer les données d'entraînement
- de gérer les joueurs mixtes (réseau + algorithme)

## Classes de gestion de l'environnement

| Classe | Fichier | Usage |
| ------ | ------- | ----- |
| Board | board.py | Représentation d'un état du jeu Onitama |
| Card | card.py | Représentation d'une carte de mouvement |
| Action | boardpy | Représentation d'une action |
| Move | card.py | Représentation d'un mouvement |

## Gestion de jeu

| Classe | Fichier | Usage |
| ------ | ------- | ----- |
| Game | game.py | Joue une partie contre P1 et P2 |
| GameSession | game.py | Joue un nombre N de parties entre P1 et P2 |

## Joueurs de référence

| Classe | Fichier | Usage |
| ------ | ------- | ----- |
| HeuristicEvaluation | heuristic.py | Permet l'évaluation heuristique (différentes méthoides) d'un état |
| Player | players.py | Classe parente de tous les joueurs |
| HApiPlayer | players.py | Joueur humain (pour l'app) |
| HumanPlayer | players.py | Joueur humain (en console) |
| HeuristicPlayer | players.py | Joueur exploitant des règles heuristiques |
| LookAheadHeuristicPlayer | players.py | Joueurs exploitant des règles heuristiques sur N niveaux |

## Génération des données d'entraînement

| Classe | Fichier | Usage |
| ------ | ------- | ----- |
| RegularDataTrainer | trainer.py | Classe de gestion du buffer pour la génération des données d'entraînement |

## Entrainement PPO
