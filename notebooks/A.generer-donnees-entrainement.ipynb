```python
import sys
sys.path.append('../onitama/')

from players import LookAheadHeuristicPlayer
from trainer import RegularDataTrainer
from game import GameSession
```

```python
dstFolder = "../data/supervised/"
```

```python
training_plan = [
    (
        LookAheadHeuristicPlayer(max_depth=2, heuristic_function="heuristic_defensive"),
        LookAheadHeuristicPlayer(max_depth=2, heuristic_function="heuristic_positional"),
        "defensive2-vs-positional2"
    )
]
```

```python
i = 1
for p1, p2, filename in training_plan:
    print(f"Training session {i} : {filename}")
    trainer = RegularDataTrainer(
        p1=p1,
        p2=p2,
        p1_record=True,
        p2_record=True,
        save_only_wins=True,
        x_file_destination=f"{dstFolder}{filename}-states.pkl",
        y_file_destination=f"{dstFolder}{filename}-actions.pkl",
        v_file_destination=f"{dstFolder}{filename}-values.pkl",
        override=True
    )
    gameSession = GameSession(player_one=p1, player_two=p2, number_of_games=1, trainer=trainer)
    gameSession.start()
```
