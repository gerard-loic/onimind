# Version des réseaux de neurones et des modèles

Le présent notebook fait l'inventaire de l'évolution de l'architecture des réseaux de neurones ainsi que des modèles entraînés

## Evolution de l'architecture des réseaux de neurones

| Version réseau | Type | Architecture | Evolution | Utilisation |
| --- | --- | --- | --- | --- |
| v1 | CNN | cf. images/onitama_network_v4_architecture.png | | Abandonné. Architecture présentant des goulots |
| v2 | CNN | cf. images/onitama_network_v4_architecture.png | Métriques additionnelles de validation k-accuracy, utilisation de AdamW au lieu de Adam pour réduire l'over fitting | Abandonné. Architecture présentant des goulots, mais améliorations AdamW et métriques k-accuracy conservées |
| v3 | CNN | cf. images/onitama_network_v4_architecture.png | A partir de v2, mais réduite | Abandonné. Architecture présentant des goulots |
| v4 | CNN | cf. images/onitama_network_v4_architecture.png | Version avec inférence tensorflow optimisée et gestion de l'entraînement PPO | Utilisé pour modèle Musashi mais résultats moyens, présence du goulot dans les couches |
| v5 | CNN | cf. images/onitama_network_v5_architecture.png | Architecture optimisée | Architecture optimisée pour corriger les problèmes de goulot, utilisé pour modèle Tokugawa |
| v6 | CNN | cf. images/onitama_network_v6_architecture.png | Implémentation de l'entraînement avec masque sur nouvelles données d'entraînement regénérées | Utilisée pour modèle Kamae, meilleur modèle CNN pur |
| v7 | Dense | cf. images/onitama_network_v7_architecture.png | Premier essai d'architecture sur base de réseau dense | Utilisé pour modèle Tairanauchu |
| v8 | CNN | cf. images/onitama_network_v8_architecture.png | Issu de l'architecture de v6, mais avec 10 couches de blocs résiduels (plus profond) | Abandonné. Trop lourd. |
| v9 | Dense | cf. images/onitama_network_v9_architecture.png | Architecture identique à v7, sur réseau dense, mais avec une couche de plus dans la tête de politique | Base du modèle Shigemori |
| v10 | Dense | cf. images/onitama_network_v10_architecture.png | Architecture sur réseau dense, sans LayerNormalization, sans Dropout, allégée avec skip connection | Base du modèle Sukoshi |
